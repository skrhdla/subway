```python
#호선과 역명별로 그룹화하고 필요없는 열 삭제
instation_group=geton_subway_passenger.groupby(['호선','역명']).sum()
instation_group=instation_group.drop(['연번'],axis=1)
#instation_group=instation_group.drop(['날짜'],axis=1)
instation_group=instation_group.drop(['역번호'],axis=1)
#instation_group=instation_group.drop(['구분'],axis=1)
instation_group=instation_group.drop(['합 계'],axis=1)
```
