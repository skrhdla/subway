```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
model_G=model.groupby('역명')['총인원'].sum().sort_values()
model_ListIndex=model_G.index.tolist()[7:]
model_List=model_G.values.tolist()[7:]
```

```python
model_Set=list(zip(model_ListIndex,model_List))
model_DF=pd.DataFrame(model_Set)
model_DF.columns=['역명','총인원',]
```

```python
plt.rc("font",family="Malgun Gothic",size=18, weight="light")
plt.figure(figsize=(50,25))
colors=['lightgray','lightgray','lightgray','lightblue','lightblue','skyblue','skyblue','royalblue','royalblue','salmon']
wedgeprops={'width':0.7,'edgecolor':'w','linewidth':5}
x=model_DF['역명']
y=model_DF['총인원']
plt.pie(y,labels=x,autopct='%d%%',counterclock=False,colors=colors,wedgeprops=wedgeprops)
plt.xlabel("역명",size=30,weight='bold')
plt.title("역별 총인원 유동인구 수",size=30,weight='bold')
plt.show()
```
